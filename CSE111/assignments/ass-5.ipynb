```python
# task3
class Spaceship:
    def __init__(self,shipname,capacity):
        self.shipname=shipname;self.capacity=capacity
        self.material_lst=[];self.mater_weight=[]
    def load_cargo(self,*material):
        for i in material:
            if i.getWeight()<self.capacity:
                self.material_lst.append(i.getName())
                self.mater_weight.append(i.getWeight())
                self.total=0
                for j in self.mater_weight:
                    self.total+=j
            else:
                print(f"Warning: Unable to load {i.getName()} inside {self.shipname}.")
                self.mater_weight.append(i.getWeight())
                self.total_value=0
                for k in self.mater_weight:
                    self.total_value+=k
                print(f'Exceeds capacity by {self.total_value-self.capacity}')
    def display_details(self):
        print(f"Spaceship Name: {self.shipname}\nCapacity: {self.capacity}\nCurrent Cargo Weight: {self.total}\nCargo: {self.material_lst}")
class Cargo:
    def __init__(self,name,weight):
        self.__name=name;self.__weight=weight
        # self.totalweight+=self.__weight
    def getName(self):return self.__name
    def getWeight(self):return self.__weight
# Creating spaceships
falcon = Spaceship("Falcon", 50000)
apollo = Spaceship("Apollo", 100000)
enterprise = Spaceship("Enterprise", 220000)
print("1.===================================")
# Creating cargo
gold = Cargo("Gold", 20000)
platinum = Cargo("Platinum", 25000)
dilithium = Cargo("Dilithium", 50000)
trilithium = Cargo("Trilithium", 70000)
neutronium = Cargo("Neutronium", 80000)
print("2.===================================")
# Loading cargo onto spaceships
falcon.load_cargo(gold)
falcon.load_cargo(platinum)
falcon.display_details()
print("3.===================================")
apollo.load_cargo(gold)  # Apollo will not reach its total capacity
apollo.display_details()
print("4.===================================")
falcon.load_cargo(neutronium)  # This should exceed Falcon's capacity
print("5.===================================")
enterprise.load_cargo(dilithium)
enterprise.load_cargo(trilithium)
enterprise.load_cargo(neutronium)  # This should not exceed Enterprise's capacity
enterprise.display_details()
```

```python
class Train:
  def __init__(self, name, *stops):
    self.name = name
    self.start = stops[0]
    self.end = stops[-1]
    self.stops = list(stops)
    self.passengers = []
    print(f"Welcome aboard on {self.name}")
    print(f"Start: {self.start}")
    print(f"Destination: {self.end}")
  def addPassenger(self, *obj):
    for i in obj:
      if i.start == None:
        i.start = self.start
      if i.end == None:
        i.end = self.end
      fare = 100 * (self.stops.index(i.end) - self.stops.index(i.start))
      self.passengers.append([i.name, i.start, i.end, fare])
      print(f"{i.name} welcome aboard")
  def allPassengerDetails(self):
    for i in self.passengers:
      print(f"Name: {i[0]}, Start: {i[1]}, Destination: {i[2]}, Fare: ${i[3]}")
class Passenger:
  def __init__(self, name, start = None, end = None):
    self.name = name
    self.start = start
    self.end = end
t1 = Train('T1-Express','New York','Manhattan','Brooklyn','Boston')
print("1========================")
p1 =Passenger("Naruto")
t1.addPassenger(p1)
p2 = Passenger("Sasuke","Manhattan")
p3 = Passenger("Hinata","Manhattan","Brooklyn")
print("2========================")
t1.addPassenger(p2,p3)
print("3========================")
t1.allPassengerDetails()
print("4========================")
t2 = Train('Europe-Express','London','Paris','Brussels','Turkey')
print("5========================")
p4 =Passenger("Max","London","Brussels")
p5 = Passenger("Eleven","Paris")
p6 = Passenger("Mike","Brussels")
t2.addPassenger(p4,p5,p6)
print("6========================")
t2.allPassengerDetails()
```
