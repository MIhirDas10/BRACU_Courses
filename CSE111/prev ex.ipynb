```python
# task3
class PizzaMachine:
    def __init__(self,name='Regular',size=6):
        self.name=name;self.size=size
        self.toppings=[];self.spice=None
    def customizePizza(self,toppings=None,spice='Regulary'):
        self.spice=spice
        if toppings==None:
            print(f"No toppings specified! Can't bake pizza.")
        else:
            allowed_spice=['Regular','Hot','Super Naga']
            if type(toppings)==list:
                for i in toppings:
                    if i not in allowed_spice:
                        print("Sorry! Spice level not allowed. Can't bake Pizza.")
                    else:
                        self.toppings.append(i)
                # print(self.toppings)
            out=', '.join(self.toppings)
        return f'Your {self.size}-inch {self.spice} spicy {self.name} Pizza is ready with {out} toppings. Enjoy!'
#Write your code here
pizza1 = PizzaMachine()
order1 = pizza1.customizePizza(["Cheese", "Pepperoni"],"Hot")
print("1################################# ")
print(order1)
print("2================================")
pizza2 = PizzaMachine("Vege")
order2 = pizza2.customizePizza("Super Naga")
print("3#################################")
print(order2)
print("4================================")
pizza3 = PizzaMachine("Chicken Blast",12)
order3 = pizza3.customizePizza(["Mushroom"])
print("5#################################")
print(order3)
print("6================================")
pizza4 = PizzaMachine("Beef Bonanza",16)
order4 = pizza4.customizePizza(["Cheese","Beef kalabhuna"],"Mild")
print("7#################################")
print(order4)
print("8================================")
```
