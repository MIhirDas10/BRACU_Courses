```python
# task4
class NikeBangladesh:
    total_status={'Air Jordan': 0, 'Cortez': 0, 'Zoom Kobe': 0}
    total_sold= 0
    branch_opened=[]
    def __init__(self,branch):
        self.branch=branch
        self.branch_status={'Air Jordan': 0, 'Cortez': 0, 'Zoom Kobe': 0}
        self.sold=0
        self.branch_opened.append(branch)
    @classmethod
    def status(cls):
        print(f'Nike Bangladesh Status:\nBranches Opened: {NikeBangladesh.branch_opened}')
        print(f'Currently Stocked\n{cls.total_status}')
        print(f'Sold: {cls.total_sold}')
    def details(self):
        print(f'Nike {self.branch} outlet:\nProducts Currently Stocked:')
        print(f'{self.branch_status}\nSold: {self.sold}')
    def restockProducts(self,dict1):
        for k,v in dict1.items():
            self.branch_status[k]+=v
            NikeBangladesh.total_status[k]+=v
    def productSold(self,dict2):
        for k,v in dict2.items():
            self.branch_status[k]-=v
            self.sold+=v
            NikeBangladesh.total_status[k]-=v
            NikeBangladesh.total_sold+=v
print("xxxxxxxxxxxxxx1xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
dhaka = NikeBangladesh("Dhaka Banani")
chittagong = NikeBangladesh("Chittagong GEC")
print("xxxxxxxxxxxxxx2xxxxxxxxxxxxxxxx")
dhaka.details()
print("xxxxxxxxxxxxxx3xxxxxxxxxxxxxxxx")
chittagong.details()
print("xxxxxxxxxxxxxx4xxxxxxxxxxxxxxxx")
dhaka.restockProducts({"Air Jordan":1200,"Cortez":200,"Zoom Kobe":200})
chittagong.restockProducts({"Air Jordan":1000,"Cortez":250,"Zoom Kobe":100})
print("xxxxxxxxxxxxxx5xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
print("xxxxxxxxxxxxxx6xxxxxxxxxxxxxxxx")
dhaka.productSold({"Air Jordan":760,"Cortez":90})
chittagong.productSold({"Air Jordan":520,"Zoom Kobe":70})
print("xxxxxxxxxxxxxx7xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
```
