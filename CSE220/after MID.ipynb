```python
from sympy import *
def extremum(f, r):
  df = diff(f, x)
  ddf = diff(df, x)
  root = solveset(df, x,r)
  roots = [i.evalf() for i in root]
  f_val = [f.subs([(x,i)]).evalf() for i in roots]

  ddf_val = [ddf.subs([(x,i)]).evalf() for i in roots]
  if len(roots) == 0:
    print("No minima or maxima")
  else:
    for i in range(len(roots)):
      if ddf_val[i] < 0:
        print(f"at x={roots[i]}, we have local maxima, f({roots[i]})= {f_val[i]}")
      elif ddf_val[i] > 0:
        print(f"at x={roots[i]}, we have local minima, f({roots[i]})= {f_val[i]}")

x=Symbol('x')
f = (sin(x)-sin(2*x))
i = Interval(0,5)
print('Extrema of the function: (minima/maxima)')
extremum(f, i)
```

```python
import numpy as np
def findMax(arr,n):
    # base case
    if n==1:
        return arr[0]
    # recursive case
    max_of_rest=findMax(arr[1:],n-1)
    # comparision
    if arr[0]>max_of_rest:
        return arr[0]
    else:
        return max_of_rest
arr=np.array([1,2,3,4,5,10,5,3,4])
result=findMax(arr,len(arr))
print(f'The maximum number is: {result}')
```

```python
def sumEvens(arr,idx=0):
    if idx==len(arr):
        return 0
    if arr[idx]>0 and arr[idx]%2==0:
        return arr[idx]+sumEvens(arr,idx+1)
    else:
        return sumEvens(arr,idx+1)
arr=np.array([1,2,3,4,5,6,8])
result=sumEvens(arr)
print(f'Sum of the even numbers: {result}')
```

```python
def alternating_harmonic_sum(n):
    if n == 1:
        return 1
    elif n % 2 == 0:
        return -1 / n + alternating_harmonic_sum(n - 1)
    else:
        return 1 / n + alternating_harmonic_sum(n - 1)

def sum(n):
    result = alternating_harmonic_sum(n)
    return round(result, 15)
print(f'Harmonic Sum(3): {alternating_harmonic_sum(3)}') #This should print 0.8333333333333333
print(f'Harmonic Sum(4): {alternating_harmonic_sum(4)}') #This should print 0.5833333333333333
```

```python
class Node:
    def __init__(self, elem, left = None, right = None):
        self.elem = elem
        self.left = left
        self.right = right
```

```python
def preOrder(root):
    # root - left - right
    if root == None:
        return

    print(root.elem, end = ", ")
    preOrder(root.left)
    preOrder(root.right)


def InOrder(root):
    # left - root - right
    if root == None:
        return

    InOrder(root.left)
    print(root.elem, end = ", ")
    InOrder(root.right)
```

```python
def createTree(arr, index = 1):
    if index >= len(arr):
        return
    if arr[index] != None:
        root = Node(arr[index])
        root.left = createTree(arr, 2*index)
        root.right = createTree(arr, 2*index+1)
        return root
r = createTree(arr)
preOrder(r)
```

```python
def calculateNode(root):
    # base case
    if root == None:
        return 0

    # recursive call
    left = calculateNode(root.left)
    right = calculateNode(root.right)
    return left + right + 1
print(calculateNode(r))
```

```python
def sumNode(root):
    if root==None:
        return 0
    left=sumNode(root.left)
    right=sumNode(root.right)
    return left+right+root.elem
print(sumNode(r))
```

```python
def evenSum(root):
    if root==None:
        return 0
    if root.elem%2==0:
        left=evenSum(root.left)
        right=evenSum(root.right)
        return left+right+root.elem 
    else:
        return evenSum(root.left)+evenSum(root.right)
print(evenSum(r))
```

```python
def oddSum(root):
    if root==None:
        return 0
    if root.elem%2!=0:
        left=oddSum(root.left)
        right=oddSum(root.right)
        return left+right+root.elem
    else:
        return oddSum(root.left)+oddSum(root.right)
print(oddSum(r))
```

```python
def diffSum(root):
    if root==None:
        return 0
    else:
        return oddSum(r)-evenSum(r)
print(diffSum(r))
```

```python
def height(root,depth=0):
    if root==None:
        return -1
    left=height(root.left,depth+1)
    right=height(root.right,depth+1)
    return max(left,right)+1
print(height(r))
```

```python
def level(root):
    if root==None:
        return 0
    else:
        return height(r)+1
print(level(r))
```

```python
# sum(even level)-sum(odd level)
def sumEvenLevel(root):
    if root==None:
        return 0
    if root.elem%2==0:
        left=level(root.left)
        right=level(root.right)
        return max(left,right)
print(sumEvenLevel(root))
```
