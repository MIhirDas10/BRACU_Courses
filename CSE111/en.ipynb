```python
class Tech:
    def __init__(self,name,type,price,old='New'):
        self.__name=name;self.type=type;self.old=old;self.price=int(price)
        self.total=0
    def getName(self):
        return self.__name
    def expense(self):
        if 10000<=self.price<=20000:
            print(f'Item: {self.getName()}')
            print('Not that expensive')
        elif 20000<=self.price<=30000:
            print(f'Item: {self.getName()}')
            print('Moderate expensive...doable')
        elif 30000<=self.price<=50000:
            print(f'Item: {self.getName()}')
            print('Pretty expensive')
        else:
            print('Very expensive')
    def condition(self):
        if self.old=='Old':
            print('Product is old. So, the condition might not be the best')
        elif self.old=='New':
            print('Product is brand new. So, the condition is also top-notch')
    def printDetail(self):
        return f'Name: {self.getName()}\nType: {self.type}\nAge of the tech: {self.old}\nPrice: {self.price}'
t1=Tech('Computer','Tech','35000')
print(t1.printDetail())
print('----------------------------------------')
t1.expense()
print('----------------------------------------')
t1.condition()
print('----------------------------------------')
print('----------------------------------------')
t2=Tech('Tv','Entertainment','20000','Old')
print(t2.printDetail())
print('----------------------------------------')
print('----------------------------------------')
t2.expense()
print('----------------------------------------')
t2.condition()




```
